```python
import pandas as pd
import numpy as np
import datetime
```

```python
url='https://raw.githubusercontent.com/a-forty-two/COG_GN22CDBDS001_MARCH_22/main/US_Crime_Rates_1960_2014.csv'
crime = pd.read_csv(url)
crime.head()
```

```python
crime.dtypes
```

```python
crime.Year = pd.to_datetime(crime.Year, format = '%Y')
crime.head()
```

```python
crime.set_index('Year', drop=True, inplace=True)
crime.head()
```

```python
crime.drop(columns='Total')
crime.head()
```

```python
crimes = crime.resample('10AS').sum()
crimes
```
